```python
# Connect to sqlite db
db_file = r'database.sqlite'
db_file
```

```python
# Connect to sqlite db
engine = create_engine(r"sqlite:///{}".format(db_file))
```

```python
#find out what tables are in the database
sql = "SELECT name FROM sqlite_master WHERE type = 'table';"

table_names = pd.read_sql(sql, engine)
table_names
```

```python
type(table_names)
```

```python
#find out what is in the tables
sql = "SELECT * FROM BoardGames;"
boardGameData = pd.read_sql(sql, engine)
boardGameData
```

```python
sql = "SELECT COUNT(*) FROM BoardGames AS Num_Of_Rows_BoardGames;"
numRowsBoardGames = pd.read_sql(sql, engine)
numRowsBoardGames
```
