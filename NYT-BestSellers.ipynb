```python
import requests
import json
import time  #library used to get current time
import pandas as pd
```

```python
#create a text file that contains your API key
#you DO NOT want others to see your API key
#this method is not the most secure but will work for this exercise

#this will read the file and store your API key as a variable
with open('NYT_Best_App.txt', 'r') as file:
    for line in file:
        api_key = line
```

```python
api_key = api_key.rstrip()
```

```python
#sandbox database to practice getting data from their API
#url = r"https://sandbox-api.brewerydb.com/v2/beers/?key="
url = r"https://api.nytimes.com/svc/books/v3/lists/current/hardcover-fiction.json?api-key="
```

```python
#in the request to the website, give it the url and add the API key at the end
response = requests.get(url+api_key)
```

```python
#status code is 200, so this is good
response
```

```python
#this is the JSON string data from the API
response.text
```

```python
#set the string data to a variable
str_books = response.text
```

```python
#load in the string data as JSON
bookdata = json.loads(str_books)
```

```python
#create empty lists to extract data

primeISBN10_ls = [] #primary ISBN10 of the book
primeISBN13_ls = [] #primary ISBN13 of the book
title_ls = [] #title of the book
author_ls = [] #author book 
rank_ls = [] #rank of book
last_weeks_rank_ls = [] #last week's rank
number_weeks_on_list_ls = [] #number of weeks on the list
book_price_ls = [] #the price of book 
book_description_ls = [] #book_description
publisher_ls = [] #publisher of the book
book_image_ls = [] #book_image of the book
```

```python
for book in books_var['books']: 
    print(book['title'])
```

```python
#set a "timer" for 3 sec after the code starts
#end = time.time() + 3 

#while the current time is less than the ending time (3 sec after the code started)
#while time.time() <= end:  
for book in books_var['books']:
    #get the value of each key and if the key doesn't exist, set a variable to be None
    try: primeISBN10 = book['primary_isbn10'] 
    except: primeISBN10 = None
    
    try: primeISBN13 = book['primary_isbn13']
    except: primeISBN13 = None
            
    try: title = book['title']
    except: title = None
            
    try: author = book['author']
    except: author = None
            
    try: rank = book['rank']
    except: rank = None
            
    try: last_weeks_rank = book['rank_last_week']
    except: last_weeks_rank = None
            
    try: number_weeks_on_list = book['weeks_on_list']
    except: number_weeks_on_list = None
            
    try: book_price = book['isbns']['price']
    except: book_price = None
        
    try: book_description = book['description']
    except: book_description = None
            
    try: publisher = book['publisher']
    except: publisher = None
        
    try: book_image = book['book_image']
    except: book_image = None        
    
    primeISBN10_ls.append(primeISBN10)
    primeISBN13_ls.append(primeISBN13)
    title_ls.append(title)
    author_ls.append(author)
    rank_ls.append(rank)
    last_weeks_rank_ls.append(last_weeks_rank)
    number_weeks_on_list_ls.append(number_weeks_on_list)
    book_price_ls.append(book_price)
    book_description_ls.append(book_description)
    publisher_ls.append(publisher)
    book_image_ls.append(book_image)
```

```python
for book in books_var['books']:
    #get the value of each key and if the key doesn't exist, set a variable to be None
    try: primeISBN10 = books_var['books'][0]['primary_isbn10'] 
    except: primeISBN10 = None
        
    primeISBN10_ls.append(primeISBN10)
```

```python
#match indices of each list and zip into one list
bookList = list(zip(primeISBN10_ls, primeISBN13_ls, title_ls, 
                    author_ls, rank_ls, last_weeks_rank_ls, number_weeks_on_list_ls, 
                    book_price_ls, book_description_ls, publisher_ls, book_image_ls))

#names for columns in data frame
colnames = ['primary ISBN10', 'primary ISBN13','title', 'author','rank', 'last weeks rank', 'number of weeks on the list',
            'price', 'description', 'publisher', 'book image']
```

```python
#create data frame with column names in it
df = pd.DataFrame(bookList, columns=colnames)

df.head(10)
```
